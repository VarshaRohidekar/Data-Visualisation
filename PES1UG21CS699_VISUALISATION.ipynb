```python
#Importing libraries required for data visualisation.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
Data = pd.read_csv('led')
print(Data.head())
```

```python
print(Data.columns)
```

```python
#Data type of all columns in a csv file
Data.dtypes
```

```python
Countries_Append=[]
countries = Data['Country'].unique()
countries.sort()
Countries_Append.append(countries)
Countries_Append
```

```python
#Correlation
Data.corr()
```

```python
''' A heatmap contains values representing various shades of the same
colour for each value to be plotted.'''
plt.figure(figsize=(50,50))
sns.heatmap(Data.corr(),annot=True,center=1)
plt.show()
```

```python
#To find outliers,we use boxplot
for DataVals in Data.columns:
    if((Data[DataVals].dtype=='int64') | (Data[DataVals].dtype=='float64')):
        sns.boxplot(Data[DataVals],palette='viridis')
        plt.show()
```

```python
#Darker colors indicate that the data points are concentrated in that region,
#while white bins indicate that there is no data.
#bins are created on the bsais of group intervals as we cannot use scatterplot for large data sets as it gives rise to overplotting
sns.jointplot(x='Life expectancy ',y='Alcohol',data=Data,kind='hex')
plt.show()
sns.jointplot(x='Life expectancy ',y='GDP',data=Data,kind='hex')
plt.show()
```

```python
for DataVals in Data.columns:
    if((Data[DataVals].dtype=='int64') | (Data[DataVals].dtype=='float64')):
        sns.distplot(Data[DataVals])
        plt.show()
```

```python
for DataVals in Data.columns:
    missing_series = pd.isnull(Data[DataVals])
Data[missing_series]
```

```python
outlier_column=[]
for DataVals in Data.columns:
    if((Data[DataVals].dtype!='object') & (len(Data[DataVals].unique())>2)):
        outlier_column.append(DataVals)
```

```python
outlier_column
```

```python
for DataVals in outlier_column:
    Q1=Data[DataVals].quantile(0.25)
    Q3=Data[DataVals].quantile(0.75)
    IQR=Q3-Q1
    upper=Q3+ 1.5*IQR
    lower=Q1- 1.5*IQR
    Data=Data[(Data[DataVals]>lower) & (Data[DataVals]<upper)]
```

```python
for DataVals in outlier_column:
    sns.boxplot(Data[DataVals],palette='magma')
    plt.show()
```
