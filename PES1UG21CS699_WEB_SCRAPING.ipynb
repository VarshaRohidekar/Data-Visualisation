```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.goodreads.com/shelf/show/book-reviews"
page = requests.get(url)
print(page.status_code)
```

```python
Soup = BeautifulSoup(page.content,"html.parser")
```

```python
print(Soup.prettify())
```

```python
book_titles = Soup.find_all(class_ = 'bookTitle')
for book in book_titles:
    print(book.text)
```

```python
book_authors = Soup.find_all(class_ = 'authorName__container')
for book in book_authors:
    print(book.text.strip())
```

```python
book_rating = Soup.find_all(class_ = 'greyText smallText')
for book in book_rating:
    print(book.get_text().split("\n                ")[1].split("—")[0][-5:-1])
```

```python
book_rating = Soup.find_all(class_ = 'greyText smallText')
for book in book_rating:
    print(book.get_text().split("\n                ")[2].split("—")[0].split()[0])
```

```python
book_rating = Soup.find_all(class_ = 'greyText smallText')
for book in book_rating:
    print(book.get_text().split("\n                ")[3].split("—")[0].split()[1])
```

```python
import pandas as pd
book_title_list = [x.get_text() for x in book_titles]
book_author_list = [x.get_text()[1:-1] for x in book_authors]
book_rating_list = [x.get_text().split("\n                ")[1].split("—")[0][-5:-1] for x in book_rating]
book_rating_count = [x.get_text().split("\n                ")[2].split("—")[0].split()[0] for x in book_rating]
book_publishing_year = [x.get_text().split("\n                ")[3].split("—")[0].split()[1] for x in book_rating]
```

```python
Data = pd.DataFrame()
Data['Book Titles'] = book_title_list
Data['Book Authors'] = book_author_list
Data['Book Rating'] = book_rating_list
Data['Number of Ratings'] = book_rating_count
Data['Year of Publishing'] = book_publishing_year
```

```python
Data
```

```python
df = pd.DataFrame(Data)
df
```

```python
df.to_csv('Books.csv',index=False)
```
